# 数据集探索

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
driver_imgs_list = pd.read_csv("driver_imgs_list.csv")
```

```python
driver_imgs_list.head()
```

# 司机类别分布

```python
plt.figure(figsize=(12,8))
sns.countplot(x="classname", data=driver_imgs_list, color='royalblue')
plt.ylabel('Count', fontsize=18)
plt.xlabel('Class', fontsize=18)
plt.ylim(0,2500)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title("Frequency of 10 classes", fontsize=22)
plt.savefig('Fig1.jpg', dpi=300)
plt.show()
```

# 司机id分布

```python
plt.figure(figsize=(16,8))
sns.countplot(x="subject", data=driver_imgs_list, color='royalblue')
plt.ylabel('Count', fontsize=18)
plt.xlabel('Class', fontsize=18)
plt.ylim(0,1250)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title("Frequency of 26 drivers", fontsize=22)
plt.savefig('Fig2.jpg', dpi=300)
plt.show()
```

# 单个司机类别分布

```python
drivers_id = list(driver_imgs_list['subject'].unique())
drivers_id
```

```python
fig, axs = plt.subplots(ncols=6, nrows=5, figsize=(30,30))

for i in range(5):
    for j in range(6):
        n = i*6+j
        if (n+1) > len(drivers_id):
            break
        driver_id = drivers_id[n]
        ax=axs[i][j]
        sns.countplot(x="classname", data=driver_imgs_list[driver_imgs_list['subject']==driver_id], 
                      color='royalblue', ax=ax)
        ax.set_ylabel('Count', fontsize=10)
        ax.set_xlabel('Class', fontsize=10)
        ax.set_title(driver_id, fontsize=20)
plt.savefig('Fig3.jpg', dpi=300)
plt.show()
```
